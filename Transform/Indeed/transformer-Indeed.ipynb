```python
import pandas as pd
```

```python
statename = 'Illinois'
region = 'Midwest'
data_analyst = "../Resources/Indeed/"+statename+"_outputData+Analyst.xlsx"
data_engineer  = "../Resources/Indeed/"+statename+"_outputData+Engineer.xlsx"
data_scientist = "../Resources/Indeed/"+statename+"_outputData+Scientist.xlsx"
```

```python
df_Analyst = pd.read_excel(data_analyst)
df_Engineer = pd.read_excel(data_engineer)
df_Scientist = pd.read_excel(data_scientist)
```

```python
df_Analyst.dropna()
df_Engineer.dropna()
df_Scientist.dropna()
```

```python
df_Analyst["title"]="Data Analyst"
df_Engineer["title"]="Data Engineer"
df_Scientist["title"]="Data Scientist"

```

```python
# filter only three most relevant position
indeed_df = pd.concat([df_Analyst,df_Engineer,df_Scientist])

indeed_df.dropna()

indeed_df
```

```python
 indeed_df["state1"]=statename
indeed_df["region"]=region
  
indeed_df
```

```python
outputFilename = "../../Load/Resources/Indeed/"+statename+"_Indeed.csv"
indeed_df.to_csv(outputFilename,sep=',',index=False)
```
